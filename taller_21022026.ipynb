```python
import random
import time

def juego_par_impar_vs_maquina():
    """
    Juego donde el usuario ingresa un n√∫mero y la m√°quina intenta
    adivinar aleatoriamente si es par o impar.
    """
    print("--- ü§ñ Juego: Desaf√≠a a la Matrix (Par o Impar) ü§ñ ---")

    # 1. Solicitar el n√∫mero al usuario de forma segura
    while True:
        try:
            entrada = input("Ingresa un n√∫mero entero para jugar: ")
            numero_usuario = int(entrada)
            break  # Si la conversi√≥n a entero funciona, salimos del bucle
        except ValueError:
            print("‚ö†Ô∏è Error: Eso no es un n√∫mero entero v√°lido. Int√©ntalo de nuevo.")

    # 2. Determinar la realidad: ¬øEl n√∫mero del usuario es par o impar?
    if numero_usuario % 2 == 0:
        realidad_numero = "par"
    else:
        realidad_numero = "impar"

    # 3. La m√°quina hace su predicci√≥n aleatoria
    print("\nLa m√°quina est√° analizando tu n√∫mero...")
    time.sleep(1) # Peque√±a pausa para dar emoci√≥n

    opciones_maquina = ["par", "impar"]
    prediccion_maquina = random.choice(opciones_maquina)

    print(f"üîÆ Predicci√≥n de la m√°quina: ¬°Tu n√∫mero es {prediccion_maquina.upper()}!")
    time.sleep(0.5)

    # 4. Mostrar la realidad y comparar
    print(f"üí° Realidad: El n√∫mero {numero_usuario} es efectivamente {realidad_numero.upper()}.")

    print("\n--- Resultado Final ---")
    # La m√°quina gana si su predicci√≥n es igual a la realidad
    if prediccion_maquina == realidad_numero:
        print("üí• ¬°La m√°quina ha acertado su predicci√≥n!")
        print("Ha ganado la Maquina, Matrix!!")
    else:
        print("üéâ ¬°La m√°quina se ha equivocado!")
        print("Le Ganaste a la Matrix!!:)")
    print("-----------------------")

# Ejecutar el juego
if __name__ == "__main__":
    juego_par_impar_vs_maquina()
```

```python
import random
import time
import os

def limpiar_pantalla():
    # Limpia la consola seg√∫n el sistema operativo
    os.system('cls' if os.name == 'nt' else 'clear')

def imprimir_encabezado():
    print("=" * 50)
    print("      üïπÔ∏è  BIENVENIDO A: PAR O IMPAR MATRIX üïπÔ∏è      ")
    print("=" * 50)

def jugar():
    limpiar_pantalla()
    imprimir_encabezado()

    # Entrada del usuario con estilo
    try:
        n_usuario = int(input("\nüëâ Ingrese un n√∫mero para desafiar al sistema: "))
    except ValueError:
        print("\n‚ùå ERROR: La Matrix solo acepta n√∫meros enteros.")
        return

    # Proceso de "Pensamiento" de la M√°quina
    print("\n[ANALIZANDO N√öMERO...]")
    time.sleep(0.8)

    prediccion_maquina = random.choice(["PAR", "IMPAR"])
    es_par = (n_usuario % 2 == 0)
    resultado_real = "PAR" if es_par else "IMPAR"

    # Visualizaci√≥n de Resultados en Formato Pro
    print("\n" + "‚îå" + "‚îÄ" * 48 + "‚îê")
    print(f"‚îÇ  TU N√öMERO: {str(n_usuario).center(34)} ‚îÇ")
    print(f"‚îÇ  PREDICCI√ìN DE LA MATRIX: {prediccion_maquina.center(21)} ‚îÇ")
    print("‚îî" + "‚îÄ" * 48 + "‚îò")

    time.sleep(1)
    print(f"\nüì¢ LA REALIDAD ES: ¬°El n√∫mero {n_usuario} es {resultado_real}! \n")

    # Verificaci√≥n de Ganador
    print("=" * 50)
    if prediccion_maquina == resultado_real:
        print("         üíÄ ¬°LA MATRIX HA GANADO! üíÄ         ")
        print("          (Su predicci√≥n fue acertada)         ")
    else:
        print("         üéâ ¬°LE GANASTE A LA MATRIX! üéâ        ")
        print("          (La m√°quina fall√≥ el c√°lculo)         ")
    print("=" * 50)

    # Opci√≥n de reintento
    volver = input("\n¬øDeseas intentar otra vez? (s/n): ").lower()
    if volver == 's':
        jugar()
    else:
        print("\nDesconectando de la Matrix... Adi√≥s. üëã")

if __name__ == "__main__":
    jugar()
```

```python
def cifrado_cesar_13(texto):
    resultado = ""
    desplazamiento = 13

    for caracter in texto:
        # Verificar si es una letra may√∫scula
        if caracter.isupper():
            # (C√≥digo ASCII - Inicio Alfabeto + Desplazamiento) % 26 + Inicio Alfabeto
            nuevo_ascii = (ord(caracter) - 65 + desplazamiento) % 26 + 65
            resultado += chr(nuevo_ascii)

        # Verificar si es una letra min√∫scula
        elif caracter.islower():
            nuevo_ascii = (ord(caracter) - 97 + desplazamiento) % 26 + 97
            resultado += chr(nuevo_ascii)

        # Si no es letra (espacio, punto, etc.), se queda igual
        else:
            resultado += caracter

    return resultado

# --- Prueba del programa ---
mensaje_original = input("üìù Escribe el mensaje a cifrar: ")
mensaje_cifrado = cifrado_cesar_13(mensaje_original)

print(f"\nüîê Mensaje Cifrado: {mensaje_cifrado}")
print(f"üîì Si lo vuelves a cifrar: {cifrado_cesar_13(mensaje_cifrado)}")
```

```python
import hashlib

mensaje_original = input("üìù Escribe el mensaje a cifrar: ")
# Encode the message to bytes before hashing
mensaje_cifrado = hashlib.md5(mensaje_original.encode()).hexdigest()

print("Original:", mensaje_original)
print("Cifrado (MD5):", mensaje_cifrado)
```

```python
import pandas as pd

ruta_archivo ="/content/compras_semana.csv"

datos = pd.read_csv(ruta_archivo)
datos ['total'] = datos['precio'] * datos['cantidad']

total_compras=datos['total'].sum()
print("total de compra", total_compras)

promedio_compras=datos['total'].mean()
print("promedio de compra", promedio_compras)

mediana_compras=datos['total'].median()
print("mediana de compra", mediana_compras)

compra_minima=datos['total'].min()
print("compra minima", compra_minima)

compra_maxima=datos['total'].max()
print("compra maxima", compra_maxima)

std=datos['total'].std()
print("desviacion estandar", std)

var=datos['total'].var()
print("varianza", var)

registro_conteo=datos['total'].count()
print("registro conteo", registro_conteo)

datos ['total'].describe()
```
